# 3.- Tabajando con Series de tiempo en Python

```python
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns
sns.set()
```

## White Noise

```python
wn=np.random.normal(loc = df.market_value.mean(), scale = df.market_value.std(), size = len (df))
```

```python
df['wn']=wn
```

```python
df.describe()
```

```python
df.wn.plot(figsize = (20,5))
plt.title('White Noise Time-Series', size = 24)
plt.show()
```

```python
df.market_value.plot(figsize = (20,5))
plt.title('S&P Prices', size = 24)
plt.ylim(0,2300)
plt.show()
```

## Random Walk

```python
rw = pd.read_csv('C:/Users/Jorge/Documents/GitHub/curso-series-temporales/3. Caracter√≠sticas fundamentales/Python/Randwalk.csv')
```

```python
rw.head()
```

```python
rw.date = pd.to_datetime(rw.date, dayfirst=True)
rw.set_index('date',inplace=True)
rw.head()
```

```python
rw=rw.asfreq('b')
rw.describe()
```

```python
df['rw']=rw.price
df.head()
```

```python
df.rw.plot(figsize = (20,5))
plt.title('Random Walk', size = 24)
plt.show()
```

```python
df.rw.plot(figsize = (20,5))
df.market_value.plot()
plt.title('Random Walk vs S&P', size = 24)
plt.legend()
plt.show()
```

```python
# Generamos 500 pasos aleatorios para la caminara, con media 0 y desv tipica 1
steps = np.random.normal(loc=0, scale=1, size=500)
 
# Ponemos el primer elemento a 0 para que el precio inicial sea el que pongamos nosotros
steps[0]=0
 
# Simulando los precios P con un precio inicial de 100
P = 100 + np.cumsum(steps)

# Plot
plt.plot(P)
plt.title("Random Walk Simulada")
plt.show()
```

## Stationarity

```python
sts.adfuller(df.market_value)
```

```python
sts.adfuller(df.wn)
```

```python
sts.adfuller(df.rw)
```

## Seasonality

```python
s_dec_additive = seasonal_decompose(df.market_value, model = 'additive')
s_dec_additive.plot()
#plt.gcf().set_size_inches(25,15)
plt.show()
```

```python
s_dec_multiplicative = seasonal_decompose(df.market_value, model = 'multiplicative')
s_dec_multiplicative.plot()
#plt.gcf().set_size_inches(25,15)
plt.show()
```

## ACF

```python
sgt.plot_acf(df.market_value, lags = 40, zero = False)
plt.title('ACF S&P', size = 24)
plt.show()
```

```python
sgt.plot_acf(df.wn, lags=40, zero=False)
plt.ylim(-0.05,0.05)
plt.title('ACF WN', size=24)
plt.show()
```

```python
sgt.plot_acf(df.rw, lags=40, zero=False)
plt.title('ACF RW', size=24)
plt.show()
```

# PACF

```python
sgt.plot_pacf(df.market_value, lags = 40, zero = False, method ='ols')
plt.title('PACF S&P', size = 24)
plt.show()
```

```python
sgt.plot_pacf(df.wn, lags = 40, zero = False, method ='ols')
plt.title('PACF S&P', size = 24)
plt.ylim(-0.06,0.06)
plt.show()
```

```python
sgt.plot_pacf(df.rw, lags = 40, zero = False, method ='ols')
plt.title('PACF S&P', size = 24)
plt.show()
```
