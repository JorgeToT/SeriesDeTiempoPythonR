```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

# Visualizando Patrones

## Datos

```python
df = pd.read_csv('C:/Users/Jorge/Documents/GitHub/curso-series-temporales/3. Características fundamentales/Python/Ti_data.csv', index_col=0, parse_dates=True)
df
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))

df.plot(ax = ax, color = 'C0', legend = None)

ax.set_ylim(17.75,22)
ax.set_ylabel('$*C$')
ax.set_title('Internal Temperature $(T_i)$', size= 20)

plt.show()
```

## Visualizando Patrones Diurnos

```python
df ['date'] = df.index.normalize()
df['time'] = df.index.time.astype(str)
df
```

```python
df_pivot = df.pivot(index='date', columns='time', values='Ti')
df_pivot
```

```python
fig, ax = plt.subplots(1,1,figsize = (10, 6))
df_pivot.T.plot (ax = ax, color = 'C0', legend = False)
ax.set_xlim(0,47)
ax.set_ylim(17.75,22)
ax.set_ylabel('$°C$')
ax.set_xlabel('Internal Temperature $(T_i)$')
plt.show()
```

## Transparencia Alpha

```python
fig, ax = plt.subplots(1,1,figsize = (10, 6))
df_pivot.T.plot (ax = ax, color = 'C0', alpha=0.1,legend = False)
ax.set_xlim(0,47)
ax.set_ylim(17.75,22)
ax.set_ylabel('$°C$')
ax.set_xlabel('Internal Temperature $(T_i)$')
plt.show()
```

## Cambiamos el Horario de Comienzo

```python
df_roll = pd.DataFrame(index=df_pivot.index, columns=np.roll(df_pivot.columns,-36), data = np.roll(df_pivot.values, -36))

fig, ax = plt.subplots(1,1,figsize = (10, 6))
df_roll.T.plot (ax = ax, color = 'C0', alpha = 0.1,legend = False)
ax.set_xlim(0,47)
ax.set_ylim(17.75,22)
ax.set_ylabel('$°C$')
ax.set_xlabel('Internal Temperature $(T_i)$')
plt.show()
```

## Visualizando Patrones Diurnos y Semanales

```python
day_colors = {
    'Monday' : 'C0',
    'Tuesday' : 'C1',
    'Wednesday' : 'C1',
    'Thursday' : 'C1',
    'Friday' : 'C1',
    'Saturday' : 'C2',
    'Sunday' : 'C3'
}

fig, ax = plt.subplots(1,1,figsize = (10, 6))

for day_name, color in day_colors.items():
    df_roll[df_roll.index.day_name() == day_name].T.plot(ax = ax, color =color, alpha=0.1, legend = None)

ax.axvline(30, alpha=0.25, color='black', linestyle ='--')
ax.axvline(40, alpha=0.25, color='black', linestyle ='--')
ax.axvline(42, alpha=0.25, color='black', linestyle ='--')

ax.set_xlim(0,47)
ax.set_ylim(17.75,22)

ax.set_ylabel('$°C$')
ax.set_xlabel('Internal Temperature $(T_i)$')

from matplotlib.lines import Line2D
custom_lines = [
    Line2D([0], [0], color='C0', lw = 2),
    Line2D([0], [0], color='C1', lw = 2),
    Line2D([0], [0], color='C2', lw = 2),
    Line2D([0], [0], color='C3', lw = 2)
]

ax.legend(custom_lines, ['Mon', 'Tue-Fri', 'Sat', 'Sun'])
plt.show()
```
